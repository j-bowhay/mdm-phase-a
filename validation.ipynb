```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import pandas as pd
%matplotlib notebook
```

```python
vector_data = np.loadtxt("vectorgeom.csv", delimiter=',')
ow_data = np.loadtxt("ow.csv", delimiter=',')
normal_data = np.loadtxt("normal.csv", delimiter=',')
```

```python
from scipy.io import netcdf_file
import os

f = netcdf_file(os.path.join(os.path.abspath('') ,"04102019.nc"))
lat = f.variables['latitude'].data
long = f.variables['longitude'].data
vel_u = f.variables['ugos'].data.squeeze()
vel_v = f.variables['vgos'].data.squeeze()

```

```python
longm, latm = np.meshgrid(long, lat)
plt.figure()

plt.plot(vector_data[:,1],vector_data[:,0],'y.', markersize=5, label='Vector Geometry')
plt.plot(ow_data[:,1],ow_data[:,0],'g.', markersize=5, label='OW')
plt.plot(normal_data[:,1],normal_data[:,0],'b.', markersize=5, label='Normal')
#plt.quiver(longm, latm, vel_u, vel_v, color='r', scale=5)
plt.gca().set_aspect('equal')
plt.legend()
```
