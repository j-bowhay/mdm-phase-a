```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import pandas as pd
%matplotlib notebook
```

```python
validation_data = np.loadtxt("validation_data/trail_validation_04102019.csv", delimiter=',')
validation_data
```

```python
vector_data = np.loadtxt("vectorgeom", delimiter=',')
vector_data
```

```python
from scipy.io import netcdf_file
import os

f = netcdf_file(os.path.join(os.path.abspath('') ,"04102019.nc"))
lat = f.variables['latitude'].data
long = f.variables['longitude'].data
vel_u = f.variables['ugos'].data.squeeze()
vel_v = f.variables['vgos'].data.squeeze()

```

```python
longm, latm = np.meshgrid(long, lat)

plt.plot(vector_data[:,1],vector_data[:,0],'y.', markersize=10, label='Vector Geometry')
plt.plot(validation_data[:,0],validation_data[:,1],'c.', markersize=10, label='Actual')
plt.quiver(longm, latm, vel_u, vel_v, color='r', scale=5)
plt.gca().set_aspect('equal')
plt.legend()
```
