```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.style.use(['science','ieee'])
xlim = (-5, 5)
ylim = (-5, 5)

x = np.linspace(*xlim, num=25)
y = np.linspace(*ylim, num=25)

xx, yy = np.meshgrid(x, y, indexing='xy')

def vort_x(x, y):
    return (-y)/(1 + x**2 + y**2)

def vort_y(x, y):
    return (x)/(1 + x**2 + y**2)

V_x = vort_x(xx,yy)
V_y = vort_y(xx,yy)
vel_magnitude = np.hypot(V_x, V_y)

plt.figure()
plt.quiver(xx, yy ,V_x, V_y, scale=15)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Velocity Field")
plt.gca().set_aspect('equal')
plt.autoscale(tight=True)
plt.savefig("figures/vfield.pdf")
```
